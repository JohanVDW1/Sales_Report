```python
 # Import all the data I will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

 # stock informarion of the NASDAQ

db = pd.read_csv("C:/Users/johan/Downloads/archive (2)/prices-split-adjusted.csv")
db
```

```python
 # Attempting to clean the data by checking for and removing null values
 # Seeing where the null values are  

db.isnull().sum()
```

```python
 # Checking the null data
    
db[db['name'].isnull()]
```

```python
 # Shape of data with null values
    
db.shape
```

```python
#Shape of data without null

db.dropna(how = 'any').shape
```

```python
 # Removing null values
    
db = db.dropna(how = 'any')
```

```python
db.shape
```

```python
sum(db.duplicated())
```

```python
 # sorting the data base alphabetically then by date of the position

db = db.sort_values(['symbol','date'])
db
```

```python
db.nunique()
```

```python
db['date'] = pd.to_datetime(db['date'])
```

```python
db.dtypes
```

```python
 # Describing the data base
    
db.describe()
```

```python
 # Pie chart for what sectors are the largest part of the NASDAQ

value = db['sector'].value_counts()
my_colors = ['darkorange', 'dimgrey','blue','green','red','yellow','pink','aqua','purple','white','brown']
my_explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.pie(value.values,labels=value.index,autopct='%1.1f%%', colors=my_colors)
plt.title('Sector Comparison')
plt.show()
```

```python
 # Average close position by sector

databasebar = db.groupby('sector')['close'].agg(['mean']).reset_index()

fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_facecolor('white')
ax1.bar(databasebar['sector'],databasebar['mean'],color='aqua')
plt.title('Sector Average Closing Position')
plt.xlabel("sector")
plt.ylabel("Closing position")
plt.show()
```

```python
 # Since the last graph is too crowded, narrowed down the search

databasebar = db.groupby('sector')['close'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(5)

fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_facecolor('white')
ax1.bar(databasebar['sector'],databasebar['mean'],color='aqua')
plt.title('5 Sectors With The Hightest Average Closing')
plt.xlabel("Sectors")
plt.ylabel("Closing position")
plt.show()
```

```python
databasebar = db.groupby('sector')['close'].agg(['mean']).sort_values(by=('mean'),ascending=True).reset_index().head(5)

fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_facecolor('white')
ax1.bar(databasebar['sector'],databasebar['mean'],color='aqua')
plt.title('5 Sectors With The Lowest Average Closing')
plt.xlabel("Sectors")
plt.ylabel("Closing position")
plt.show()
```

```python
databasebar = db.groupby('sector')['volume'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(5)

fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_facecolor('white')
ax1.bar(databasebar['sector'],databasebar['mean'],color='aqua')
plt.title('5 sectors With The Hightest Average Volume')
plt.xlabel("Companies")
plt.ylabel("Volume")
plt.show()
```

```python
 # Scatter graph showing the correlation between time and close position

plt.scatter(db['date'],db['close'])

plt.xlabel("Date")
plt.ylabel("Close Position")
plt.title("The Relationship Between Date and Close Position")
plt.show()
```

```python
databasebar = db.groupby('name')['volume'].agg(['mean']).sort_values(by=('mean'),ascending=False).reset_index().head(5)

fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_facecolor('white')
ax1.bar(databasebar['name'],databasebar['mean'],color='aqua')
plt.title('5 Companies With The Hightest Average Volume')
plt.xlabel("Companies")
plt.ylabel("Volume")
plt.show()
```
